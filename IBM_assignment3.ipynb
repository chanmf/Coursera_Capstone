```python
import requests
page = requests.get("https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M")
page
```

```python
print(page.content)
```

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(page.content, 'lxml')
```

```python
table = soup.find('div', class_ = 'mw-parser-output').table
print(table.prettify())
```

```python
table
```

Create pandas dataframe with the data from the table of wikipedia page.

```python
data = []
```

```python
columns = ['PostalCode', 'Borough', 'Neighbourhood']
```

```python
for tr in table.find_all('tr'):
    try:
        templist = []
        i = 0
        for td in tr.find_all('td'):
            templist.append(td.text.strip())
            i += 1
            if i == 3:
                data.append(templist)
                del templist
                i = 0
    except Exception as e:
        pass
    
```

```python
data
```

```python
import pandas as pd
df = pd.DataFrame(data, columns = columns)
df
```

Drop all the cells with *Borough* that is **Not assigned**.

```python
df[df['Borough'] == 'Not assigned'].index
```

```python
df.drop(df[df['Borough'] == 'Not assigned'].index, inplace = True)
df
```

```python
ind = df[df['Neighbourhood'] == 'Not assigned'].index
```

```python
df.loc[ind,'Borough'].values
```

```python
df.loc[ind,'Neighbourhood'] = df.loc[ind,'Borough'].values
```

```python
df
```

Combine *Neighbourhood* separated with comma for **repeated** postal code area

```python
df2 = df[df['PostalCode'].duplicated() == True]
df2
```

```python
df.drop(df2.index, inplace = True)
df
```

```python
for index, data2 in df2.iterrows():
    for index, data in df.iterrows():
        if data2['PostalCode'] == data['PostalCode']:
            data['Neighbourhood'] = data['Neighbourhood'] + ', ' + data2['Neighbourhood']
```

```python
df
```

Merge the dataframe with *geographical coordinates* for each postal code

```python
dfx = pd.read_csv('Geospatial_Coordinates.csv')
dfx
```

```python
dfx.rename(columns = {'Postal Code':'PostalCode'}, inplace = True)
```

```python
df_new = pd.merge(df, dfx, on = 'PostalCode')
```

```python
df_new
```

```python
from geopy.geocoders import Nominatim

geolocator = Nominatim(user_agent = 'trt_explorer')
location = geolocator.geocode('Toronto')
latitude = location.latitude
longitude = location.longitude
print('The geographical coordinate of Toronto are {}, {}'.format(latitude, longitude))
```

To generate map view with all postal code marking. 

```python
import folium

map_toronto = folium.Map(location = [latitude, longitude], zoom_start = 10)

for lat, lng, label in zip(df_new['Latitude'], df_new['Longitude'], df_new['Borough']):
    folium.CircleMarker (
        [lat,lng],
        radius = 5,
        popup = label,
        color = 'blue',
        fill = True,
        fill_color='#3186cc',
        fill_opacity=0.7,
    ).add_to(map_toronto)
    
map_toronto
```

To generate map that group the markers which are closed together.

```python
from folium.plugins import MarkerCluster

toronto_map = folium.Map(location = [latitude, longitude], zoom_start = 10)

cluster = MarkerCluster().add_to(toronto_map)

for lat, lng, label in zip(df_new['Latitude'], df_new['Longitude'], df_new['Neighbourhood']):
    folium.Marker(
        location = [lat, lng],
        icon = None,
        popup = label,
    ).add_to(cluster)
    
toronto_map
```

The map generated shows the quantity of postal code near Downtown Toronto is larger and concentrated while the other areas have lesser and scattered.
